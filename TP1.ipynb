<h3>Execise 1</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# generate a random data
ID = 210055
seed = ID % 10000
np.random.seed(seed)
x = 3 * np.random.rand(120,1)
y = 1.8 + 2.5 * x + np.random.randn(120,1)
```

```python
# ploting the random generated data
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Ploting")
plt.show()
```

```python
# spliting data (70/30)
x_train , x_test,y_train , y_test = train_test_split(x , y , test_size= 0.3 , random_state = 611)
x_test = np.array(x_test).reshape(-1,1)
y_test = np.array(y_test)
x_train = np.array(x_train).reshape(-1,1)
y_train = np.array(y_train)
```

```python
# creating LinearRegression Model
model = LinearRegression()
# use the trainig data to train the model
model.fit(x_train , y_train)
# printing Coef and Intercept , y = 2.41x + 1.74
print("w = " , model.coef_)
print("b = " , model.intercept_)
```

```python
# using the score function
score = model.score(x_test , y_test)
print("score on test set was: ",score)
```

```python
# calculate the predicted values of x_test
y_p = model.predict(x_test)
# ploting the data beside the regeression line
plt.scatter(x_test , y_test , color = 'blue' , label = 'True date')
plt.plot(x_test , y_p , color ='red', label = 'Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# generated random new data
x_new = np.array([[0.5] ,[1.5] , [2.5]])
x_new = x_new.reshape(-1,1)
# predection its y values based on the regression model
y_new = model.predict(x_new)
# ploting them next to the original data set (testing data set)
plt.scatter(x_test , y_test , color = 'blue' , label = 'True date')
plt.scatter(x_new , y_new , color ='red', label = 'new data predected')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
